# Exploratory Data Analysis


```python

import pandas as pd

fear_df = pd.read_csv('fear_greed_index.csv')
```

```python

# Create the 'date' column from 'timestamp'
fear_df['date'] = pd.to_datetime(fear_df['timestamp'], unit='s')

# Set 'date' as the index
fear_df.set_index('date', inplace=True)

# Check the result
print(fear_df.head())
print(fear_df.columns.tolist())  # should now show: ['timestamp', 'value', 'classification']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(fear_df.index, fear_df['value'], color='orange', marker='o')
plt.title('Fear-Greed Index Over Time')
plt.xlabel('Date')
plt.ylabel('Index Value (0 = Fear, 100 = Greed)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The Fear-Greed Index over the years (2018–2025) shows frequent and sharp fluctuations, indicating that market sentiment shifted rapidly between fear and greed.
Major dips reflect high market uncertainty or panic, likely during global crises (e.g., 2020).
Peaks near 80–90 suggest periods of overconfidence, while recent years (2023–2025) indicate a moderate recovery in investor sentiment, hovering between 40–70.

```python
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.countplot(x='classification', data=fear_df, palette='Set2')
plt.title('Frequency of Market Sentiment Categories')
plt.xlabel('Sentiment')
plt.ylabel('Number of Days')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Out of all recorded days, Fear was the most dominant market sentiment, followed by Greed and Extreme Fear.
This suggests that traders mostly operated in a cautious or uncertain environment.
Extreme Greed days were the least frequent, indicating that highly optimistic sentiment was rare during this period.
The overall distribution reflects a bearish market trend with short-term recoveries.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='classification', y='value', data=fear_df, palette='coolwarm')
plt.title('Index Value Distribution by Sentiment Class')
plt.ylabel('Fear-Greed Index')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The boxplot of the Fear-Greed Index reveals that most values fall within the 20–70 range, with a median around 40–50, showing that market sentiment leaned slightly toward caution.
Several outliers below 10 and above 90 indicate rare but extreme emotions, i.e., panic crashes or euphoric rallies.
The presence of these extremes suggests emotional trading behavior among market participants, which can impact short-term price volatility.

## CONCLUSION
The Fear-Greed Index analysis from 2018 to 2025 reveals that markets have mostly operated under cautious or fearful conditions, with occasional spikes of optimism. Fear was the most common sentiment, suggesting uncertainty and risk-averse behavior among traders. Extreme Greed was rare, indicating that overly confident market phases did not last long. The presence of frequent ups and downs, along with emotional extremes, shows that market sentiment has been unstable, influenced by major global events and investor psychology. Overall, traders mostly acted cautiously, and emotional factors had a strong influence on market trends.
